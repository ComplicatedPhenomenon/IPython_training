```python
from __future__ import print_function

# Import conventions we'll be using here. See Part 1
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
datafile = open('DATAgg4.txt','r')
data =  datafile.readlines() # Python File readlines() Method
datafile.close()

for row in data:
    this_data = row.split()

# print (type(this_data))

S = []
cs = []
for row in data:
    this_data = row.split()
    S.append(float(this_data[0]))
    cs.append(float(this_data[1]))

#print ('\n')
#print (type(S))

S = np.array([float(row.split()[0]) for row in data])
CS = np.array([float(row.split()[1]) for row in data])
#print (S)
#print (CS)
fig,ax = plt.subplots()
ax.plot(S,CS,label='.') # turn pb into fb
#plt.legend()
#plt.title(r'$qq\longrightarrow t\bar{t}G_n$', fontsize=20)
ax.set( xlabel='m')
ax.legend()
plt.show()
```

Having `fig` is useful if you want to change figure-level attributes or save the figure as an image file later (e.g. with fig.savefig('yourfilename.png'). You certainly don't have to use the returned figure object but many people do use it later so it's common to see.
