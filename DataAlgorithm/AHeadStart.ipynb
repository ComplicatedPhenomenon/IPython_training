```md
Input: [1,8,6,2,5,4,8,3,7]
Output: 49
```

````python
possibleVolume = []
input = [1,8,6,2,5,4,8,3,7]
for i in range(len(input)):
    for j in range(i+1, len(input)):
        possibleVolume.append((j-i)*min([input[i], input[j]]))
````

````python
max(possibleVolume)
````
