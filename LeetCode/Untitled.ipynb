```python
def sortedSquares( A):
    return list(sorted([a ** 2 for a in A]))
```

```python
sortedSquares([-4,-1,0,3,10])
```

```python
def largestPerimeter( A):
    A.sort()
    n = len(A)
    for i in range(1, len(A) - 1):
        if A[n - i - 2] + A[n - i - 1] > A[n - i]:
            return A[n - i - 2] + A[n - i - 1] + A[n - i]
    return 0
```

```python
largestPerimeter([-4,-1,1,3,10,13])
```
